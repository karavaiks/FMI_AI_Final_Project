Обучение нейронной сети - полносвязная нейронная сеть с тремя слоями

```python
#преобразуем целевые переменные в массив numpy
y_train = y_train.to_numpy() if isinstance(y_train, pd.Series) else y_train
y_test = y_test.to_numpy() if isinstance(y_test, pd.Series) else y_test

#создание тензоров из данных для использования в PyTorch
#тензоры - структуры данных, как нампай массивы, они поддерживают вычисления на ГПУ
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)
```

```python
#создание датасетов из признаков и меток
#объединение тестового тензора и целевых меток - теперь датасет состоит из пар (признак, метка)
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
#загрузчик данных, разбивая датасет на мини-батчи по 32 штуки, перемешиваем
#нужно для ускорения обучения по порциям 
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
#определение модели бинарной классификации
class BinaryClassificationModel(nn.Module):
    def __init__(self, input_size):
        super(BinaryClassificationModel, self).__init__()
        self.fc1 = nn.Linear(input_size, 64) #первый полносвязный слой
        self.relu = nn.ReLU() #функция активации ReLu
        self.dropout = nn.Dropout(0.3) #Dropout для регуляризации
        self.fc2 = nn.Linear(64, 32) #второй полносвязный слой
        self.fc3 = nn.Linear(32, 1) #выходной слой 
        self.sigmoid = nn.Sigmoid() #сигмоидальная функция активации для вероятностей
    
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.sigmoid(self.fc3(x))
        return x
```

```python
#создание модели с количеством входных признаков
input_size = X_train.shape[1]
model = BinaryClassificationModel(input_size)
```

```python
#определение функции потерь - бинарная кросс-энтропия и оптимизатора - Adam
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
#обучение модели
epochs = 50 #количество эпох
for epoch in range(epochs):
    model.train() #режим обучения
    for batch_X, batch_y in train_loader:
        optimizer.zero_grad() #обнуляем градиенты
        outputs = model(batch_X) #прямой проход
        loss = criterion(outputs, batch_y) #вычисляем ошибку
        loss.backward() #обратное распространение ошибки
        optimizer.step() #обновление параметров модели
    print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}') #выводим ошибку
```

```python
model.eval() #переключаем модель в режим оценки
predictions = []
actuals = []
with torch.no_grad(): #открлючаем градиенты
    for batch_X, batch_y in test_loader:
        outputs = model(batch_X) #прогоняем тестовые данные через модель 
        predicted = (outputs > 0.5).float() #преобразуем вероятности в классы
        predictions.extend(predicted.numpy().flatten()) #сохраняем предсказания
        actuals.extend(batch_y.numpy().flatten()) #сохраняем реальные значения
```

```python
accuracy = accuracy_score(actuals, predictions)
precision = precision_score(actuals, predictions)
recall = recall_score(actuals, predictions)
f1 = f1_score(actuals, predictions)
conf_matrix = confusion_matrix(actuals, predictions)
report = classification_report(actuals, predictions)
```

```python
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(report)
```

```python
# ROC-кривая
fpr, tpr, _ = roc_curve(actuals, predictions)
plt.figure()
plt.plot(fpr, tpr, color='blue', label='ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
